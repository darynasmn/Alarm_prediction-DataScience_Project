```python
import pandas as pd
from IPython.core.display import display, HTML
from bs4 import BeautifulSoup
```

```python
INPUT_DATA_FOLDER = '1_isw_prepared'
DATA_FILE = 'isw_all_days.csv'
OUTPUT_FOLDER = '2_isw_parsed'
OUTPUT_DATA_FILE = 'isw_all_days_parsed.csv'
```

```python
df = pd.read_csv(f'{INPUT_DATA_FOLDER}/{DATA_FILE}', sep=';')
```

```python
df.head()
```

```python
test_row = df.iloc[10]
page_html_text = test_row['main_html']
print(page_html_text)
```

```python
display(HTML(page_html_text))
```

```python
def remove_names_and_dates(page_html_text):
    parsed_html = BeautifulSoup(page_html_text)
    p_lines = parsed_html.findAll('p')
    
    min_sentences_word_count = 20
    p_index=0
    
    for p_line in p_lines:
        strong_lines = p_line.findAll('strong')
        if not strong_lines:
            p_index += 1
            continue
            
        for s in strong_lines:
            if len(s.text.split(' '))>= min_sentences_word_count:
                break
        else:
            p_index +=1
            continue
        break
    for i in range(0, p_index):
        page_html_text = page_html_text.replace(str(p_lines[i]), '')
    return page_html_text
```

```python
df['main_html_v2'] = df['main_html'].apply(lambda x: remove_names_and_dates(x))
```

```python
import re

pattern = '\[(\d+)\]'

df['main_html_v3'] = df['main_html_v2'].apply(lambda x: re.sub(pattern, '', x))
```

```python
test_row = df.iloc[10]
page_html_text_v4 = test_row['main_html_v3']
display(HTML(page_html_text_v4))
```

```python
df['main_html_v4'] = df['main_html_v3'].apply(lambda x: BeautifulSoup(x).text)
```

```python
df.head(10)
```

```python
df['main_html_v5'] = df['main_html_v4'].apply(lambda x: re.sub(r'http(\S+.*\s)','',x))
df['main_html_v6'] = df['main_html_v5'].apply(lambda x: re.sub(r'(o2022|o2023|2022|2023)','',x))
```

```python
df.tail()
```

```python
df['main_html'] = df['main_html'].apply(lambda x: BeautifulSoup(x).text)
df = df.drop(['main_html_v3','main_html_v4',
               'main_html_v5'], axis = 1)
```

```python
df.tail()
```

```python
df.to_csv(f'{OUTPUT_FOLDER}/{OUTPUT_DATA_FILE}', sep =';', index = False)
```
